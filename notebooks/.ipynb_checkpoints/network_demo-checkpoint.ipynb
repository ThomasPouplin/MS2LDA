A function to extract a networkx graph containing only the motifs we're interested in

```python
def get_motif_graph(ms2lda, motifs_of_interest):

    G, json_data = ms2lda.get_network_graph(to_highlight=None, degree_filter=0)
    motifs_of_interest = ['motif_' + str(id) for id in motifs_of_interest]

    # 1. keep only the motifs in the list
    remove_count = 0
    nodes = G.nodes(data=True)
    for node_id, node_data in nodes:
        # 1 for doc, 2 for motif
        if node_data['group'] == 2 and node_data['name'] not in motifs_of_interest: 
            remove_count += 1
            G.remove_node(node_id)
    print "Removed %d motifs from the graph because they're not in the list" % remove_count

    # 2. keep only motifs having shared nodes with other motifs
    removed_motifs = []
    nodes = G.nodes(data=True)    
    for node_id, node_data in nodes:
        if node_data['group'] == 2:
            # check if any doc in this motif is shared with another motif (degree > 1)
            neighbours = G.neighbors(node_id)
            share = False
            for nb in neighbours:
                deg = G.degree(nb)
                if deg > 1:
                    share = True
            # if not then delete from the graph too
            if not share:
                removed_motifs.append(node_data['name'])
                G.remove_node(node_id)
    print "Removed %s from the graph because they don't share documents with other motifs in the list" % removed_motifs
    
    # 3. delete all unconnected nodes from the graph
    unconnected = []
    nodes = G.nodes(data=True)    
    for node_id, node_data in nodes:
        if G.degree(node_id) == 0:
            unconnected.append(node_id)
    G.remove_nodes_from(unconnected)   
    print "Removed %d unconnected documents from the graph" % len(unconnected)
    
    # for easy layout purpose
    motif_nodes = []
    doc_nodes = []
    doc_degrees = []
    nodes = G.nodes(data=True)    
    for node_id, node_data in nodes:
        if node_data['group'] == 2: # is a motif
            motif_nodes.append(node_id)
        elif node_data['group'] == 1: # is a doc
            doc_degrees.append(G.degree(node_id))
            doc_nodes.append(node_id)
                
    return G, doc_nodes, motif_nodes
```

So, when we create the graph, we also filtered it in such a way that only the motifs in the provided list are retained and additionally, these have to be the ones with have some documents shared with another motif in the list. 

Function to extract subgraph where we only keep only the nodes above min_degree. Also creates the bipartite layout.

```python
def extract_subgraph(G, all_doc_nodes, min_degree, spacing_left=5, spacing_right=10):

    # extract subgraph of docs > min_degree and all the connected topics
    doc_nodes_to_keep = set()
    motif_nodes_to_keep = set()
    for n in all_doc_nodes:
        if G.degree(n) >= min_degree:
            neighbours = G.neighbors(n)
            doc_nodes_to_keep.add(n)
            motif_nodes_to_keep.update(neighbours)
    to_keep = doc_nodes_to_keep | motif_nodes_to_keep # set union
    SG = G.subgraph(to_keep)
    nodes = SG.nodes(data=True)    
        
    pos = dict()
    pos.update( (n, (1, i*spacing_left)) for i, n in enumerate(doc_nodes_to_keep) ) # put doc_nodes on the left side
    pos.update( (n, (2, i*spacing_right)) for i, n in enumerate(motif_nodes_to_keep) ) # put motif_nodes on the right side
        
    # set the node and edge labels
    doc_labels = {}
    motif_labels = {}
    for node_id, node_data in SG.nodes(data=True):
        if node_data['group'] == 2: # is a motif
            motif_labels[node_id] = node_data['name']
        elif node_data['group'] == 1: # is a doc
            doc_labels[node_id] = node_data['name']        
        
    return SG, pos, doc_labels, motif_labels
```

```python
fig_width = 10
fig_height = 20

plt.figure(figsize=(fig_width, fig_height))
plt.axis('off')
fig = plt.figure(1)

min_degree = 2
SG, pos, doc_labels, motif_labels = extract_subgraph(G, all_doc_nodes, min_degree)

nx.draw_networkx_nodes(SG, pos, alpha=0.1)
nx.draw_networkx_edges(SG, pos, alpha=0.1)
_ = nx.draw_networkx_labels(SG, pos, doc_labels, font_size=10)
_ = nx.draw_networkx_labels(SG, pos, motif_labels, font_size=14)
```
