The main results from inference on the LDA model are **(1)** the document-topic matrix, which tells us the distribution of topics over **all documents** in the corpus and **(2)** the topic-word matrix, which tells us the distribution of topics over **all words** in the vocabulary. We can access them as below:

```python
doc_topic = ms2lda.model.doc_topic_
topic_word = ms2lda.model.topic_word_
```

doctopic is 1588 x 300, which is the no. of documents X no. of topics.

topicword is 300 x 3171, which is the no. of topics X no. of words.

<h1>3. Current Visualisation</h1>

So above section 1 & 2 show how you can load the results from an LDA analysis and extract useful information from them. This should contain all the information you need when building the visualisation interface (preferably as a proper standalone web application in Django).

In our current codes, we have developed a basic visualisation to allow users to explore the results in an interactive manner. First we start by thresholding the document-topic and topic-word distributions by some values. This way, we can say which topics are used in which documents, and which words 'belongs' to a topic. 

<h4>Set Visualisation Parameters</h4>

```python
# If True, an interactive visualisation is shown in a separate tab. 
# You need to interrupt the kernel to stop it once you're done with it (from the menu above, Kernel > Interrupt).
interactive=True
```

```python
# Used for highlighting 'consistent' topic fragments/losses during plotting. 
# A value of 0.50 means the fragment (or loss) word occurs at least half of the plotted documents of the topic.
consistency=0.0
```

```python
# Used for graph visualisation in the interactive mode only. 
# Specifies the 'special' nodes to be coloured differently.
# special_nodes = [
#     ('doc_372.18877_540.996', '#CC0000'), # maroon
#     ('doc_291.66504_547_239', 'gold'),
#     ('doc_308.17029_289.13', 'green'),
#     ('topic_244', '#CC0000'), # maroon
#     ('topic_98', 'aqua'),
#     ('topic_293', '#ff1493') # deep pink
# ]
special_nodes = None
```

<h4>Run Visualisation</h4>

```python
# interrupt the kernel to terminate the visualisation
ms2lda.plot_lda_fragments(consistency=consistency, interactive=interactive, to_highlight=special_nodes)
```
